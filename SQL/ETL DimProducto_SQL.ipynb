### Creación de Tablas

En el siguiente ejemplo se muestra como crear una tabla en Azure Databricks

```
%sql
CREATE TABLE Categoria
(
Cod_Categoria STRING,Nombre STRING
)
USING CSV
OPTIONS (header 'true', inferSchema 'true', delimiter ',')
LOCATION "/mnt/landing/adventuresales/sales_system/dbAventureWorks/Categoria/Categoria.csv"
```

```
%sql
SELECT * FROM Categoria
```

```
%sql
CREATE TABLE SubCategoria
(
Cod_SubCategoria STRING ,Nombre STRING ,Cod_Categoria STRING
)
USING CSV
OPTIONS (header 'true', inferSchema 'true', delimiter ',')
LOCATION "/mnt/landing/adventuresales/sales_system/dbAventureWorks/Subcategoria/SubCategoria.csv"
```

```
%sql
SELECT * FROM SubCategoria
```

```
%sql
CREATE TABLE Producto
(
Cod_Producto STRING,Nombre STRING,Cod_SubCategoria STRING, Color STRING
)
USING CSV
OPTIONS (header 'true', inferSchema 'true', delimiter '|')
LOCATION "/mnt/landing/adventuresales/sales_system/dbAventureWorks/Producto/Producto.csv"
```

```
%sql
SELECT * FROM Producto
```

```
%sql
CREATE TABLE DimProducto
AS
SELECT P.Cod_Producto,P.Nombre as Producto,S.Nombre as Subcategoria, C.Nombre as Categoria 
FROM Producto P
JOIN Subcategoria S ON P.Cod_SubCategoria=S.Cod_SubCategoria
JOIN Categoria C ON C.Cod_Categoria=S.Cod_Categoria;
```

```
%sql
SELECT * FROM DimProducto
```

### Guardado en el Data Lake

En el siguiente ejemplo se muestra como guardar un resultado al Data Lake Store

```
df = spark.table('DimProducto')
df.coalesce(1).write.format("com.databricks.spark.csv").mode ("overwrite").option("header", "true").save("/mnt/staging/adventure_sales/sales_system/dbAdventureWorks/DimProducto")
```
